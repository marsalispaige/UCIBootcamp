```python
import pandas as pd
#sqlalchemy 
from sqlalchemy import create_engine, inspect
from sqlalchemy.orm import Session

# Import and establish Base for which classes will be constructed 
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float, DateTime

import datetime
```

```python
stationdf = pd.read_csv("output/clean_hawaii_stations.csv")
measuredf = pd.read_csv("output/clean_hawaii_measurements.csv")
measuredf.head()
```

```python
#establishing connection to DB
engine = create_engine("sqlite:///hawaii.sqlite")
Base.metadata.create_all(engine)
```

```python
# Define a stations class
class station(Base):
    __tablename__ = "stations"
    
    id = Column(Integer, primary_key=True)
    station = Column(String)
    name = Column(String)
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Integer)
```

```python
# Define a stations class
class measurements(Base):
    __tablename__ = "measurements"
    
    id = Column(Integer, primary_key=True)
    station = Column(String)
    date = Column(DateTime)
    prcp = Column(Integer)
    tobs = Column(Float)
```

```python
#adding tables to hawaii DB
conn = engine.connect()
Base.metadata.create_all(conn)
```

```python
# Create the inspector and connect it to the engine
inspector = inspect(engine)
```

```python
# Collect the names of tables within the database
inspector.get_table_names()
```

```python
#print the column names within the 'measurements' table and its types
columns = inspector.get_columns('measurements')
for column in columns:
    print(column["name"], column["type"])
```

```python
#print the column names within the 'stations' table and its types
columns = inspector.get_columns('stations')
for column in columns:
    print(column["name"], column["type"])
```

```python
#establishing session for commits and query
session = Session(bind=engine)
```

```python
# adding data to the stations table with the hawaii DB
for index, row in stationdf.iterrows():
    stationadd = station(station=row["station"], name=row['name'], latitude=row["latitude"], longitude=row["longitude"],elevation=row["elevation"])
    session.add(stationadd)
    session.commit()
```

```python
#checking the data commit to DB
station_list = session.query(station)
for x in station_list:
    print(x.station)
    print(x.name)
    print(x.latitude)
    print(x.longitude)
    print(x.elevation)
```

```python
# adding data to the measurements table with the hawaii DB
for i, row in measuredf.iterrows():
    measureadd = measurements(
                            station=row["station"], 
                              date=pd.to_datetime(row['date'],format='%Y-%m-%d'), 
                              prcp=row["prcp"], 
                              tobs=row["tobs"])
    session.add(measureadd)
    session.commit()
```

```python
#checking the data commit to DB
measurement_list = session.query(measurements)
for x in measurement_list:
    print(x.station)
    print(x.date)
    print(x.prcp)
    print(x.tobs)
```
