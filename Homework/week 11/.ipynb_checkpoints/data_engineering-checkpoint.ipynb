```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading in CSV files from resources
df_measures = pd.read_csv("resources/hawaii_measurements.csv")
df_stations = pd.read_csv("resources/hawaii_stations.csv")
```

```python
#displaying dataframe attributes
df_measures.info()
print("-"*24)
df_stations.info()
```

```python
#Removing null rows and displaying dataframe attributes
df_measures = df_measures.dropna(how='any')
df_measures.info()
df_measures.head(25)
```

```python
#left join on df_measures to join df_stations location and elevation
df_hawaii_climate = df_measures.merge(df_stations,on='station',how='left')
df_hawaii_climate.info()
df_hawaii_climate.head()
```

```python
#exporting merge DF to CSV
df_measures.to_csv("output/clean_hawaii_measures.csv")
df_stations.to_csv("output/clean_hawaii_stations.csv")
df_hawaii_climate.to_csv('output/clean_hawaii_climate.csv')
```
